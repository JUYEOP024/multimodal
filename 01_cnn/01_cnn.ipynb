# CNN 개요

합성곱 신경망(Convolutional Neural Network, CNN)은 주로 이미지와 같은 2차원 데이터 분석에 적합한 신경망 구조이다.

합성곱과 풀링 계층을 사용하여 특징을 추출하고, 최종적으로 완전 연결 계층을 통해 분류 작업을 수행한다.

CNN은 이미지의 공간적 구조를 유지하면서 각 픽셀 간의 관계를 고려할 수 있는 특성이 있어, 이미지 인식, 영상 분석, 그리고 자연어 처리에서도 널리 활용된다.

1. 합성곱 층(Convolutional Layer)
    - **합성곱 연산**을 통해 이미지의 중요한 특징을 추출한다. 작은 필터(또는 커널)를 사용하여 이미지 전체를 슬라이딩하며 각 영역의 특징을 추출한다.
    - 필터는 각각 특정 패턴(예: 가장자리, 텍스처 등)을 감지하는데, 이 필터들을 학습하면서 이미지의 중요한 정보가 추출된다.
    - 각 필터의 이동 결과를 **특징 맵(feature map)**이라고 하며, 이 특징 맵들이 합쳐져서 이미지의 중요한 정보를 단계적으로 추출한다.

    - **활성화 함수**
        - 합성곱 층의 출력을 비선형 함수인 **ReLU** 함수에 통과시켜 비선형성을 부여한다. 이를 통해 네트워크가 더 복잡한 패턴을 학습할 수 있게 된다.

    - 풀링 층(Pooling Layer)
        - **맥스 풀링(Max Pooling)**이나 **평균 풀링(Average Pooling)**과 같은 연산을 통해 특징 맵의 크기를 줄여 데이터의 차원을 줄이고 계산량을 줄인다.
        - 일반적으로 가장 높은 값을 취하는 맥스 풀링이 자주 사용되며, 이를 통해 이미지의 주요 특징이 강조된다.
        - 풀링은 데이터의 위치에 대한 불변성을 높이는 데 도움을 주며, 과적합을 방지하는 효과도 있다.

2. 완전 연결 층(Fully Connected Layer)
    - 마지막으로, 추출된 특징을 활용하여 분류 작업을 수행하기 위해 **완전 연결 층**을 사용한다.
    - CNN의 마지막 단계에서 1차원 벡터 형태로 평탄화(Flatten)된 후, 완전 연결 층으로 연결되어 각 클래스에 대한 확률값을 출력한다.
  
3. CNN의 학습
    - CNN은 데이터의 특징을 학습하며 필터와 가중치를 최적화해나간다. 이를 위해 **역전파(backpropagation)**와 **경사하강법(gradient descent)**을 사용하여 필터의 파라미터를 갱신한다.
    - CNN은 이미지에서 가장자리, 질감, 모양 등의 낮은 수준의 특징을 추출한 후 점차 높은 수준의 의미 있는 특징을 추출해 최종적으로 분류를 가능하게 한다.

## Convolution Layer 주요개념

**Feature Extraction**: 이미지의 특정 패턴이나 특징을 추출하는 과정으로, CNN에서 필터와 합성곱 연산을 통해 이미지의 중요한 특징을 뽑아낸다.

**Filter (Kernel)**: 작은 행렬로, 이미지의 특정 패턴을 인식하는 역할을 한다. 필터를 이미지에 적용해 특징 맵(feature map)을 생성한다.

**필터를 한번 사용할때, 하나의 출력값이 나온다.**

### **Stride**

필터를 적용할 때 이미지 위에서 움직이는 간격이다. 스트라이드가 커질수록 **출력 크기는 줄어든다**.


![https://towardsdatascience.com/intuitively-understanding-convolutions-for-deep-learning-1f6f42faee1
](https://miro.medium.com/v2/resize:fit:588/format:webp/1*BMngs93_rm2_BpJFH2mS0Q.gif)


### **Padding**

입력 이미지 주변에 여백을 추가하는 것으로, 출력 크기를 조절하고 경계 부분의 정보 손실을 막는 데 사용된다.

- **Valid Padding**: 패딩을 주지 않는 방식으로, 출력 크기는 입력보다 작아진다.(기본값)
- **Same Padding**: 출력 크기가 입력과 동일하도록 패딩을 적용한다.

### **Pooling(Sub-sampling)**

특징 맵에서 공간 크기를 줄이고 계산량을 줄이는 과정이다. 최대 풀링(max pooling) 또는 평균 풀링(avg pooling)을 사용한다.

# Conv layer + Pooling layer

- H와 W는 이미지나 데이터의 **크기(차원)**를 나타내는 약어입니다.
    - H (Height): 이미지의 높이 (세로 길이)
    - W (Width): 이미지의 너비 (가로 길이)

```python
# Pytorch CNN 기본 흐름 : Conv2d -> ReLU -> MaxPool2d
import torch                # 텐서 생성/연산
import torch.nn as nn       # 신경망 레이어 모듈
import torch.nn.functional as F     # 활성화/풀링 등 함수형 API

input_tensor = torch.randn(1, 1, 28, 28)    # BCHW

# 합성곱 레이어 정의
conv_layer = nn.Conv2d(
    in_channels = 1,       # 입력 채널 수 (1:흑백, 3:JPG, 4:PNG)
    out_channels = 4,      # 출력 채널 수 (필터/특징맵 개수)
    kernel_size = 3,       # 3X3 채널
    stride = 1,            # 한 칸씩 이동
    padding = 'same'       # 입력, H, W == 출력 H, W이 되도록 패딩  
)

x = F.relu(conv_layer(input_tensor))       # Conv 결과에 ReLU 적용
print(x.shape)

# pool layer
x = F.max_pool2d(x, kernel_size=2, stride=2)    # 2X2 맥스 풀링(이동2)로 H, W를 절반으로 다운 샘플링
print(x.shape)
```

```python
# 이미지 전처리 : 평균 0, 표준편차 1 범위로 변환

train_images = train_dataset.data.unsqueeze(1) / 255.0  # (N, 28, 28) -> (N, 1, 28, 28) 채널 추가 + 0~1변환
train_images = (train_images - 0.5) / 0.5               # 0~1 값 분포를 -1~1로 변환

test_images = test_dataset.data.unsqueeze(1) / 255.0
test_images = (test_images - 0.5) / 0.5

train_labels = train_dataset.targets        # 학습 라벨 텐서
test_labels = test_dataset.targets          

print(train_images.shape)   # 전처리된 학습 이미지 shape
train_images[0][0]          # 첫번째 이미지의 0번째 채널 값 확인
```

train_images = (train_images - 0.5) / 0.5
- 이미지가 0~1 범위에 있고, 평균이 대략적으로 0.3~0.6 근처라서 0.5 기준으로 이동을 시키면 대략적으로 평균 0 근처로 맞춰지는 효과가있다.
- CNN에서 -1~1 범위로 맞추는 것은 학습 안정성이 좋아지는 효과가 있다. (빠른 안정성을 위한 범위 스케일링)

```python
from sklearn.model_selection import train_test_split        # 데이터 분할 함수

tr_idx, val_idx = train_test_split(
    range(len(train_images)),   # 전체 인덱스
    test_size= 0.15,            # 15%는 검증용
    random_state = 42           # 시드 고정 (재현성)
)

tr_images = train_images[tr_idx]    # 학습 인덱스에 해당하는 이미지
tr_labels = train_labels[tr_idx]    # 학습 인덱스에 해당하는 라벨
val_images = train_images[val_idx]
val_labels = train_labels[val_idx]


print(tr_images.shape, tr_labels.shape)
```

```python
# FashionNMIST 분류용 CNN 모델 정의
import torch
import torch.nn as nn 
import torch.optim as optim
import torch.nn.functional as F
from torchsummary import summary    # 모델 요약 출력 도구

# FashionNMIST 분류용 CNN 모델 클래스
class FashionMNISTCNN(nn.Module):
    def __init__(self):
        super().__init__()

        # Convolution Layer (Feature Extraction 특징 추출)
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)       # 1채널 -> 64채널, same_padding (28x28은 유지)
        self.conv2 = nn.Conv2d(64, 32, kernel_size=3, stride=1, padding=0)      # 64채널 -> 32채널, valid_padding (26x26)
        self.pool = nn.MaxPool2d(kernel_size = 2, stride=2)     # 2x2 맥스풀링으로 다운 샘플링 (13x13)

        # Fully Connected Layer (분류기)
        # - 3차원으로 되어있는 Feature Map 결과를 Fully Connected 연결하기 위해서 Flatten()을 적용해야 한다.
        self.flatten = nn.Flatten()    # (C: 32, H: 13, W: 13) -> (C*H*W)로 펼침
        self.fc1 = nn.Linear(32 * 13 * 13, 100)   # (C*H*W) -> (100)
        self.fc2 = nn.Linear(100, 10)             # (100) -> (10) 클래스 로짓 
    
    def forward(self, x):
        x = F.relu(self.conv1(x))     
        x = F.relu(self.conv2(x))
        x = self.pool(x)

        x = self.flatten(x)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x        # (batch, 10) 반환

model = FashionMNISTCNN()
summary(model, (1, 28, 28), device = 'cpu')
```

```python
# Pytorch CNN 학습 파이프라인 : Device 설정 -> DataLoader -> Train/Validation Loop -> History 기록
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader   # 텐서 데이터셋 / 배치 로더

# 디바이스 / 손실함수 / 최적화함수 설정
device = torch.device('mps' if torch.mps.is_available() else 'cpu')

model = model.to(device)
criterion = nn.CrossEntropyLoss()    # 다중분류 손실함수 (로짓 입력 + 정수 라벨)
optimizer = optim.AdamW(model.parameters(), lr = 0.001)

# 데이터로더 생성
batch_size = 128
train_dataset = TensorDataset(tr_images, tr_labels)     # (이미지, 라벨) 텐서 묶기
val_dataset = TensorDataset(val_images, val_labels)

train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle=True)    # 학습 로더는 섞어서 배치 생성
val_loader = DataLoader(val_dataset, batch_size = batch_size, shuffle=False)        # 검증 로더 순서는 고정

history = {'loss' : [], 'accuracy' : [], 'val_loss' : [], 'val_accuracy' : []}

epochs = 30
for epoch in range(epochs):
    # 학습 모드
    model.train()    # 드롭아웃, 배치 정규화 활용
    train_loss = 0.0
    train_correct = 0
    train_total = 0

    for batch_idxm, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)

        optimizer.zero_grad()   # 이전 배치 grafient 초기화
        outputs = model(data)   # 순전파 (로짓 출력)
        loss = criterion(outputs, target)   # 손실 계산
        loss.backward()     # 역전파(gradient 계산)
        optimizer.step()    # 파라미터 업데이트

        train_loss += loss.item()    # 배치 손실 누적 (스칼라)
        _, predicted = torch.max(outputs.data, 1)    # 가장 큰 로짓의 클래스 인덱스
        train_total += target.size(0)    # 배치 샘플 수 누적
        train_correct += (predicted == target).sum().item()   # 맞춘 개수 누적

    # 검증 모드
    model.eval()      # 드롭아웃, 배치 정규화 비활성화
    val_loss = 0.0
    val_correct = 0
    val_total = 0

    with torch.no_grad():    # 검증 시 gradient 계산 비활성화 (속도 / 메모리 절약)
        for data, target in val_loader:
            data, target = data.to(device), target.to(device)

            outputs = model(data)       # 순전파
            loss = criterion(outputs, target)    # 손실 계산

            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            val_total += target.size(0)
            val_correct += (predicted == target).sum().item()

    # 평균 계산
    train_loss /= len(train_loader)     # 학습 손실
    val_loss /= len(val_loader)
    train_accuracy = train_correct / train_total   # 학습 정확도
    val_accuracy = val_correct / val_total

    history['loss'].append(train_loss)
    history['accuracy'].append(train_accuracy)
    history['val_loss'].append(val_loss)
    history['val_accuracy'].append(val_accuracy)

    # 5 에폭마다 결과 출력
    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch+1}/{epochs}]')
        print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.4f}')
        print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.4f}')
        print('-'*50)

print('학습 완료!')
```

```python
# 학습 곡선 시각화 함수
import matplotlib.pyplot as plt

# history 딕셔너리를 받아서 손실/정확도 곡선을 그리는 함수
def show_history(history):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history['loss'], label='train')
    plt.plot(history['val_loss'], label='validation')
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['accuracy'], label='train')
    plt.plot(history['val_accuracy'], label='validation')
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

show_history(history)
```

```python
# 특성맵 추출
# - Pytorch에서 중간 레이어의 출력을 얻기 위한 Hook 함수

# 특정 레이어의 forward 출력(activation)을 저장하는 클래스
class FeatureExtractor:
    def __init__(self, model, layer_name):
        self.model = model              # 대상 모델
        self.layer_name = layer_name    # 추출할 레이어 이름
        self.features = None            # 추출된 출력(특성맵) 저장용
        
        # 해당 레이어의 forward hook 등록 (순전파 시 출력 가로채기)
        layer = dict(model.named_modules())[layer_name]
        layer.register_forward_hook(self.hook)   # forward 떄마다 self.hook이 호출되도록 연결

    def hook(self, module, input, output):
        self.features = output   # 해당 레이어의 출력 텐서를 저장

    def get_features(self, x):
        self.model.eval()
        with torch.no_grad():
            _ = self.model(x)     # 순전파 실행 (중간 레이어에서 hook이 features 저장)
        return self.features      # 저장된 특성 반환
```

- 지정한 layer_name 레이어의 forward 출력 (특성맵 / activation)이 self.features에 저장되고, get_features(x) 호출시 그 특성맵 텐서를 반환
- register_forward_hook(callback) : 모델의 특정 레이어가 forward 연산(순전파)을 할때, 출력값(actication 등)을 가로채거나 기록하고 싶을때 사용하는 매서드

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import cv2      # OpenCV 라이브러리 : 이미지/영상 로딩, 변환, 리사이즈, 필터링, 윤곽선, 카메라/비디오 입출력 등 제공

def show_images(images, labels, ncols=8):
    figure, axs = plt.subplots(figsize=(22, 6), nrows=1, ncols=ncols)    # 1행 ncols개의 출력 영역 생성
    for i in range(ncols):
        # PyTorch 텐서를 numpy로 변환하고 채널 순서를 변경 (C, H, W) -> (H, W, C) : imshow 출력형태
        img = images[i].permute(1, 2, 0)    # 순서 변경 함수(permute) - 채널 축을 마지막으로 이동
        # 정규화 해제 ([-1, 1] -> [0, 1])
        img = img * 0.5 + 0.5               # 원래 0~1 범위로 복원
        img = torch.clamp(img, 0, 1)        # 0 ~ 1 범위로 값 clamp (범위를 벗어나면 잘라냄)
        axs[i].imshow(img.cpu().numpy())    # CPU로 이동 후 numpy로 변환해서 출력

        # 라벨 처리
        if torch.is_tensor(labels[i]):     # 라벨이 텐서면
            label = labels[i].item()       # 파이썬 정수형으로 변환
        else:
            label = labels[i]              # 이미 정수면 그대로 사용
        axs[i].set_title(class_names[int(label)])  # 제목 정해줌
        axs[i].axis('off')                 # 축 숨김

    plt.show()

# 시각화를 위한 데이터 가져오기
vis_loader = DataLoader(trainset, batch_size=16, shuffle=False)
vis_images, vis_labels = next(iter(vis_loader))

show_images(vis_images[:8], vis_labels[:8], ncols=8)    # 8개 뿌려줌
show_images(vis_images[8:16], vis_labels[8:16], ncols=8)
```

```python
class CIFAR10_CNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CIFAR10_CNN, self).__init__()

        # 첫 번째 컨볼루션 블록 : (3, 32, 32) -> (32, 32, 32) -> pool -> (32, 16, 16)
        self.conv1_1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)   # 3채널(RGB) -> 32채널 분할, H/W는 동일
        self.conv1_2 = nn.Conv2d(32, 32, kernel_size=3, padding=1)  # 32채널 -> 32채널 (특징 강화)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)          # H, W 절반 다운샘플링

        # 두 번째 컨볼루션 블록 : (32, 16, 16) -> (64, 16, 16) -> pool -> (64, 8, 8)
        self.conv2_1 = nn.Conv2d(32, 64, kernel_size=3, padding=1)  # 32특성 -> 64채널로 분할, H/W는 동일
        self.conv2_2 = nn.Conv2d(64, 64, kernel_size=3, padding=1)  # 64채널 -> 64채널 (특징 강화)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)          # H, W 절반 다운샘플링

        # 세 번째 컨볼루션 블록 : (64, 8, 8) -> (128, 8, 8) -> pool -> (128, 4, 4)
        self.conv3_1 = nn.Conv2d(64, 128, kernel_size=3, padding=1) # 64특성 -> 128채널로 분할, H/W는 동일
        self.conv3_2 = nn.Conv2d(128, 128, kernel_size=3, padding=1) # 128채널 -> 128채널 (특징 강화)
        self.pool3 = nn.MaxPool2d(kernel_size=2, stride=2)           # H, W 절반 다운샘플링

        # 완전 연결 레이어 
        self.flatten = nn.Flatten()  # (C, H, W) -> (C*H*W)
        self.dropout1 = nn.Dropout(0.5)     # 드롭아웃 50% (과적합 완화)
        self.fc1 = nn.Linear(128 * 4 * 4, 300)  # 4x4는 32x32 이미지가 3번의 pooling을 거쳐서 나온 크기
        self.dropout2 = nn.Dropout(0.3)
        self.fc2 = nn.Linear(300, num_classes)  # 최종 로짓 출력 (클래스 수)

    def forward(self, x):
        # 첫 번째 컨볼루션 블록
        x = F.relu(self.conv1_1(x))
        x = F.relu(self.conv1_2(x))
        x = self.pool1(x)

        # 두 번째 컨볼루션 블록
        x = F.relu(self.conv2_1(x))
        x = F.relu(self.conv2_2(x))
        x = self.pool2(x)

        # 세 번째 컨볼루션 블록
        x = F.relu(self.conv3_1(x))
        x = F.relu(self.conv3_2(x))
        x = self.pool3(x)

        # 완전 연결 레이어
        x = self.flatten(x)
        x = self.dropout1(x)
        x = F.relu(self.fc1(x))
        x = self.dropout2(x)
        x = self.fc2(x)  # CrossEntropyLoss에서 softmax가 자동으로 적용됨

        return x     # (batch, num_classes) 반환

# 모델 생성
model = CIFAR10_CNN(num_classes=10)

device = torch.device('mps' if torch.mps.is_available() else 'cpu')
model = model.to(device)

# 모델 구조 출력
print(model)
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torch.optim as optim

# optimizer는 Adam으로 설정하고, label값이 원-핫 인코딩이 아니므로 loss는 CrossEntropyLoss 임.
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 학습률 스케줄러 설정 (선택사항)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)


# 검증 데이터셋 생성 (train 데이터의 15%를 검증용으로 사용)
train_size = int(0.85 * len(trainset))
val_size = len(trainset) - train_size
train_dataset, val_dataset = random_split(trainset, [train_size, val_size])

# 데이터 로더 생성
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

print(f"학습 데이터 크기: {len(train_dataset)}")
print(f"검증 데이터 크기: {len(val_dataset)}")
print(f"테스트 데이터 크기: {len(testset)}")

# 학습 함수 정의
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=30, patience=5):
    train_losses = []
    train_accuracies = []
    val_losses = []
    val_accuracies = []

    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in range(num_epochs):
        # 학습 모드
        model.train()
        train_loss = 0.0
        train_correct = 0
        train_total = 0

        for batch_idx, (data, target) in enumerate(train_loader):
            data, target = data.to(device), target.to(device)

            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, target)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            train_total += target.size(0)
            train_correct += (predicted == target).sum().item()

        # 검증 모드
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for data, target in val_loader:
                data, target = data.to(device), target.to(device)
                outputs = model(data)
                loss = criterion(outputs, target)

                val_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                val_total += target.size(0)
                val_correct += (predicted == target).sum().item()

        # 평균 계산
        train_loss = train_loss / len(train_loader)
        val_loss = val_loss / len(val_loader)
        train_acc = train_correct / train_total
        val_acc = val_correct / val_total

        train_losses.append(train_loss)
        train_accuracies.append(train_acc)
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

        print(f'Epoch {epoch+1}/{num_epochs}:')
        print(f'  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}')
        print(f'  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}')

        # 학습률 스케줄러 업데이트
        scheduler.step(val_loss)

        # 조기 종료 체크
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            # 최고 모델 저장
            torch.save(model.state_dict(), 'cifar10_best_model.pth')
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

    # 최고 모델 로드
    model.load_state_dict(torch.load('cifar10_best_model.pth'))

    return {
        'train_loss': train_losses,
        'train_accuracy': train_accuracies,
        'val_loss': val_losses,
        'val_accuracy': val_accuracies
    }

# 학습 실행
history = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=30, patience=5)
```
